## Python - List

### 基本定義

```py
# 一個 Python 的 List
[1, 2, 3, 4, 5]

# 這也是
[1, 3455, 2.45, "Happy Pollaao", False]

# 空 List
[]
```

Python 中的 List 有以下幾個特性：

- 可以放入不限數量的元素(element)
- 放入的型別不限

### 使用方法

1. 取用 list 裡的值

```
datas = [1, 2, 3, 56, 213]

# 取值
print(datas[0])      # 1
print(datas[4])      # 213
```

放在中括弧的數字叫做**index**，範圍是 0 ~ list 的長度-1，注意喔，**是從0開始喔**。

Python 比較特殊的是，index 可以為負數，比如說 index -1 就是得到最後一個值，-2 是得到倒數第二個值，其餘以此類推

2. 得到 List 的長度

```py
print(len(datas))     # 5
```

[Built-in Function - len](https://docs.python.org/3/library/functions.html#len)

3. List 相關操作

```py
# 新增元素到 list 裡
datas.append(12)

# 把元素刪除
datas.pop()     # 把最後一個元素刪除
datas.pop(1)    # 把第 2 個元素（index 為 1）的元素刪除
```

[List 的更多操作](https://docs.python.org/3/tutorial/datastructures.html)
