## How to fetch input from user?

在 Python 中，要從使用者拿到資料的話，可以使用下面的 Function:

```py
data = input("Prompt Message...: ")
```

這時候螢幕上會多出一個輸入匡，使用者可以打字進去。

注意喔，因為從 `input()` 得回來的資料，也就是存進 data 中的值的型別是字串(`str`)，所以如果要讀取數字的時候要做型別轉換：

```py
# 輸入數字
data = int(input("Prompt Message...: "))
```

`int()` 是 Python 中把值轉成數字的函式

相關連結：

- [Built-in Function: Input](https://docs.python.org/3/library/functions.html#input)
- [Built-in Function: Int](https://docs.python.org/3/library/functions.html#int)
